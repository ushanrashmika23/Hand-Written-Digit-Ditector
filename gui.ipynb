```python
import tkinter as tk
from PIL import Image, ImageTk, ImageDraw
import cv2
import numpy as np
import tensorflow as tf

model = tf.keras.models.load_model("hwDigits.h5")
win = tk.Tk()
w, h = 500, 500
bw = 7
fontButton = "Helavetica 20 bold"
count = 0


def eventFunc(e):
    canvas.create_oval(e.x-20, e.y-20, e.x+20, e.y+20, fill="black")
    imgDraw.ellipse((e.x-20, e.y-20, e.x+20, e.y+20), fill="white")


def save():
    global count

    imgArray = np.array(img)
    imgArray = cv2.resize(imgArray, (8, 8))

    cv2.imwrite("data/"+str(count)+".jpg", imgArray)
    count += 1


def clear():

    global img, imgDraw

    canvas.delete("all")
    img = Image.new("RGB", (w, h), (0, 0, 0))
    imgDraw = ImageDraw.Draw(img)


def predict():
    imgArray = np.array(img)
    imgArray = cv2.cvtColor(imgArray, cv2.COLOR_BGR2GRAY)
    imgArray = cv2.resize(imgArray, (8, 8))
    imgArray = imgArray.reshape(1, 64)
    imgArray = (imgArray/255)*15
    result = model.predict(imgArray)
    labelResult.config(text="Predicted Digit: "+str(np.argmax(result)))


canvas = tk.Canvas(win, bg="white", width=w, height=h, background="white")
canvas.grid(row=0, column=0, columnspan=4)

buttonSave = tk.Button(win, text="save", width=bw,
                       bg="green", fg="white", font=fontButton, command=save)
buttonSave.grid(row=1, column=0)

buttonPredict = tk.Button(win, text="predict", width=bw,
                          bg="blue", fg="white", font=fontButton, command=predict)
buttonPredict.grid(row=1, column=1)

buttonClear = tk.Button(win, text="clear", width=bw,
                        bg="orange", fg="white", font=fontButton, command=clear)
buttonClear.grid(row=1, column=2)

buttonExit = tk.Button(win, text="exit", width=bw, bg="red",
                       fg="white", font=fontButton, command=win.destroy)
buttonExit.grid(row=1, column=3)

labelResult = tk.Label(win, text="Predicted Digit: NON", font=fontButton)
labelResult.grid(row=2, column=0, columnspan=4)

canvas.bind('<B1-Motion>', eventFunc)

img = Image.new("RGB", (w, h), (0, 0, 0))
imgDraw = ImageDraw.Draw(img)


win.mainloop()
```
